```python
import numpy,pandas
from scipy.ndimage import convolve
import os,cv2

shape,color=(512,512),cv2.COLOR_BGR2GRAY #COLOR_BGR2HSV

def LoG_filter(image, sigma, size=None):
    if size is None:
        size = int(6 * sigma + 1) if sigma >= 1 else 7
    if size % 2 == 0:
        size += 1
    x, y = numpy.meshgrid(numpy.arange(-size//2+1, size//2+1), numpy.arange(-size//2+1, size//2+1))
    kernel = -(1/(numpy.pi * sigma**4)) * (1 - ((x**2 + y**2) / (2 * sigma**2))) * numpy.exp(-(x**2 + y**2) / (2 * sigma**2))
    kernel = kernel / numpy.sum(numpy.abs(kernel))
    result = convolve(image, kernel)
    return result

train_info=pandas.read_csv("/kaggle/input/cidaut-ai-fake-scene-classification-2024/train.csv")

if os.path.exists(f"/kaggle/working/train/")==False:
    os.mkdir(f"/kaggle/working/train/")
    os.mkdir(f"/kaggle/working/train/real/")
    os.mkdir(f"/kaggle/working/train/editada/")

for i in zip(train_info['image'],train_info['label'],range(len(train_info['image']))):
    
    image=cv2.cvtColor(cv2.resize(cv2.imread(f"/kaggle/input/cidaut-ai-fake-scene-classification-2024/Train/{i[0]}"),shape), color)

    if i[1]=="real":
        cv2.imwrite(f"/kaggle/working/train/real/{i[0]}", LoG_filter(image, 1.0))# cv2.cvtColor(image, color))
    elif i[1]=="editada":
        cv2.imwrite(f"/kaggle/working/train/editada/{i[0]}", LoG_filter(image, 1.0))# cv2.cvtColor(image, color))
```

```python
train_info=os.listdir("/kaggle/input/cidaut-ai-fake-scene-classification-2024/Test")

if os.path.exists(f"/kaggle/working/test/")==False:
    os.mkdir(f"/kaggle/working/test/")

for i in train_info:
    
    image=cv2.cvtColor(cv2.resize(cv2.imread(f"/kaggle/input/cidaut-ai-fake-scene-classification-2024/Test/{i}"),shape), color)
    cv2.imwrite(f"/kaggle/working/test/{i}", LoG_filter(image, 1.0))# cv2.cvtColor(image, color))
```
