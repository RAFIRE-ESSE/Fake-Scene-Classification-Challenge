```python
import numpy,pandas
import os,cv2

shape,color=(256,256),cv2.COLOR_BGR2LAB

train_info=pandas.read_csv("/kaggle/input/cidaut-ai-fake-scene-classification-2024/train.csv")

if os.path.exists(f"/kaggle/working/train/")==False:
    os.mkdir(f"/kaggle/working/train/")
    os.mkdir(f"/kaggle/working/train/real/")
    os.mkdir(f"/kaggle/working/train/editada/")

for i in zip(train_info['image'],train_info['label'],range(len(train_info['image']))):
    
    image=cv2.resize(cv2.imread(f"/kaggle/input/cidaut-ai-fake-scene-classification-2024/Train/{i[0]}"),shape)

    if i[1]=="real":
        cv2.imwrite(f"/kaggle/working/train/real/{i[0]}", image)#cv2.cvtColor(image, color))
    elif i[1]=="editada":
        cv2.imwrite(f"/kaggle/working/train/editada/{i[0]}", image)#cv2.cvtColor(image, color))
```

```python
train_info=os.listdir("/kaggle/input/cidaut-ai-fake-scene-classification-2024/Test")

if os.path.exists(f"/kaggle/working/test/")==False:
    os.mkdir(f"/kaggle/working/test/")

for i in train_info:
    
    image=cv2.resize(cv2.imread(f"/kaggle/input/cidaut-ai-fake-scene-classification-2024/Test/{i}"),shape)
    cv2.imwrite(f"/kaggle/working/test/{i}", cv2.cvtColor(image, color))
```
